```python
from ultralytics import YOLO
from PIL import Image

import matplotlib.pyplot as plt
```

```python
model = YOLO('yolov8n.pt')
```

```python
image_path = 'test.jpg'
results = model(image_path)

result = results[0]
im_array = result.plot()
im = Image.fromarray(im_array[..., ::-1])
plt.imshow(im)
plt.show()
```
